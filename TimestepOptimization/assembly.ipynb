```python
import numpy as np
from tqdm.notebook import tqdm
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.font_manager as font_manager 
from matplotlib.pyplot import gca
```

```python
setups = ["34", "43", "52", "25"]
colormap = dict(zip(setups, ["r", "g", "b", "k"]))
```

```python
plt.figure(figsize=(16,16))
timesteps = np.linspace(0.1, 0.6, 25)
for setup in setups:
    losses = np.load(f'../Data/Timesteps/LOSSES_{setup}.npy')
    vars = np.load(f'../Data/Timesteps/VARS_{setup}.npy')
    mean = np.array(losses)
    std = np.array(vars)
    plt.plot(timesteps, mean, colormap[setup])
    plt.fill_between(timesteps, mean - std, mean + std, colormap[setup], alpha=0.2)
    plt.plot(timesteps, mean - std, colormap[setup] + '--')
    plt.plot(timesteps, mean + std, colormap[setup] + '--')
```
