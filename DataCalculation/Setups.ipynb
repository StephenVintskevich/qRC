```python
import numpy as np
from tqdm.notebook import tqdm
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.font_manager as font_manager
from matplotlib.pyplot import gca
```

```python
from qrc.pipeline import DefaultBridgePipeline, NoPurificationPipeline
```

```python
dp = 0
signals = np.load('../Data/Signals/BINARY_SIGNALS.npy')
```

```python
# # 2 + 5
# n_qubitsA = 2
# n_qubitsB = 5
# timestep = 0.39
```

```python
# 5 + 2
n_qubitsA = 5
n_qubitsB = 2
timestep = 0.11
```

```python
pipeline = DefaultBridgePipeline(n_qubitsA, n_qubitsB, timestep)
results = []
for i in tqdm(range(100)):
    res = pipeline.evaluate(signals[i], depolarization_prob=dp)
    results.append(res)
np.save(f'../Data/Results/ALL_LEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy', np.array(results))
np.save(f'../Data/Results/CROSSLEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy',
        np.array(results)[:, :, :, n_qubitsA:])
```

```python
pipeline = NoPurificationPipeline(n_qubitsA, n_qubitsB, timestep)
results = []
for i in tqdm(range(100)):
    res = pipeline.evaluate(signals[i], depolarization_prob=dp)
    results.append(res)
np.save(f'../Data/NoPurification/ALL_LEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy', np.array(results))
np.save(f'../Data/NoPurification/CROSSLEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy',
        np.array(results)[:, :, :, n_qubitsA:])
```
