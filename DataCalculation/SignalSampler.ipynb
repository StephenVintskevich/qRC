```python
signals = []
for i in tqdm(range(100)):
    signal = np.random.choice(2, size=1200)
    signals.append(signal)
np.save('/content/gdrive/My Drive/QRC/ArticleProduction/Data/Signals/BINARY_SIGNALS.npy', np.array(signals))
```

```python
signals = []
seeds = []
narma_depth = 20
for i in tqdm(range(100)):
    seed = np.random.uniform(0, 0.2, 1200)
    signal = create_narma(narma_depth, seed)
    seeds.append(seed)
    signals.append(signal)
np.save(f'/content/gdrive/My Drive/QRC/ArticleProduction/Data/Signals/NARMA{narma_depth}_SEEDS.npy', np.array(seeds))
np.save(f'/content/gdrive/My Drive/QRC/ArticleProduction/Data/Signals/NARMA{narma_depth}_SIGNALS.npy', np.array(signals))
```
