```python
import numpy as np
from tqdm.notebook import tqdm
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.font_manager as font_manager
from matplotlib.pyplot import gca
```

```python
from qrc.pipeline import DefaultBridgePipeline, NoPurificationPipeline
```

```python
n_qubitsA = 3
n_qubitsB = 4
timestep = 0.37
```

```python
signals = np.load('/content/gdrive/My Drive/QRC/ArticleProduction/Data/Signals/BINARY_SIGNALS.npy')
```

```python
# dp = 0.001
# dp = 0.01
# dp = 0.1
dp = 0.5
```

```python
pipeline = DefaultBridgePipeline(n_qubitsA, n_qubitsB, timestep)
results = []
for i in tqdm(range(100)):
    res = pipeline.evaluate(signals[i], depolarization_prob=dp)
    results.append(res)
np.save(f'/content/gdrive/My Drive/QRC/ArticleProduction/Data/Results/ALL_LEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy', np.array(results))
np.save(f'/content/gdrive/My Drive/QRC/ArticleProduction/Data/Results/CROSSLEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy',
        np.array(results)[:, :, :, n_qubitsA:])
```

```python
pipeline = NoPurificationPipeline(n_qubitsA, n_qubitsB, timestep)
results = []
for i in tqdm(range(100)):
    res = pipeline.evaluate(signals[i], depolarization_prob=dp)
    results.append(res)
np.save(f'/content/gdrive/My Drive/QRC/ArticleProduction/Data/NoPurification/ALL_LEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy', np.array(results))
np.save(f'/content/gdrive/My Drive/QRC/ArticleProduction/Data/NoPurification/CROSSLEARNING_{n_qubitsA}{n_qubitsB}_depol{dp}.npy',
        np.array(results)[:, :, :, n_qubitsA:])
```
